```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
from datetime import datetime
```

```python
df_tesla = pdr.DataReader('TSLA', data_source='stooq')
df_tesla = df_tesla.sort_index()

df_tesla
```

```python
df_tesla['High'].plot(xlim=['2020-10-26','2025-10-22'] ,ylim=[0,900], figsize=(12,4),ls='--',c='green')
```

```python
df_tesla.index
```

```python
index=df_tesla.loc['2020-10-26':'2025-10-22'].index
share_open=df_tesla.loc['2020-10-26':'2025-10-22']['Open']
share_open
```

```python
index
```

```python
figure,axis=plt.subplots()
plt.tight_layout()
## Preventing overlapping
figure.autofmt_xdate()
axis.plot(index,share_open)
```

```python
df_tesla.resample(rule='A').min() #resample('A') → “A” মানে Annual (বছর শেষে)
```

```python
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
df_tesla.resample(rule='BA').max()
```

```python
df_tesla.resample(rule='BQS').max() #BQS → Business Quarter Start (প্রতি ত্রৈমাসিকের শুরু)
```

```python
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```
