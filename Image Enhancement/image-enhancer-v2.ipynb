```python
import cv2
from cv2 import dnn_superres
import matplotlib.pyplot as plt
import os

sr = dnn_superres.DnnSuperResImpl_create()

model_path = '/kaggle/input/8x/tensorflow1/default/1/LapSRN_x8.pb' 

sr.setModel('lapsrn', 8)  

image_path = '/kaggle/input/expariment1/WhatsApp Image 2025-05-08 at 2.00.42 PM.jpeg'  
image = cv2.imread(image_path)

# Check if image was loaded properly
if image is None:
    raise ValueError("Image not found or path is incorrect.")

upscaled = sr.upsample(image)

cv2.imwrite('/kaggle/working/upscaled.png', upscaled)

# For comparison: Bicubic resize
bicubic = cv2.resize(image, (upscaled.shape[1], upscaled.shape[0]), interpolation=cv2.INTER_CUBIC)
cv2.imwrite('/kaggle/working/bicubic.png', bicubic)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(upscaled, cv2.COLOR_BGR2RGB))
plt.title("LapSRN Upscaled")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(bicubic, cv2.COLOR_BGR2RGB))
plt.title("Bicubic Upscaled")
plt.axis('off')

plt.tight_layout()
plt.show()
```
